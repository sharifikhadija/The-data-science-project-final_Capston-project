### Task 1




##### Display the names of the unique launch sites  in the space mission


```python
!pip install pandasql

import pandasql as ps
```

```python
# Define your SQL query
query = """
SELECT DISTINCT "LaunchSite"
FROM df
"""

# Execute the query
unique_sites = ps.sqldf(query, locals())

# Display the result
print(unique_sites)
```


### Task 2


#####  Display 5 records where launch sites begin with the string 'CCA' 


```python
import pandasql as ps

# SQL query to get 5 records where LaunchSite starts with 'CCA'
query = """
SELECT *
FROM df
WHERE "LaunchSite" LIKE 'CCA%'
LIMIT 5;
"""

result = ps.sqldf(query, locals())
print(result)
```

### Task 3




##### Display the total payload mass carried by boosters launched by NASA (CRS)


```python
print(df.columns)
```

```python
# Filter rows where Customer is NASA (CRS)
nasa_crs_payload = df[df['Customer'] == 'NASA (CRS)']['PAYLOAD_MASS__KG_'].sum()

print("Total payload mass for NASA (CRS) missions:", nasa_crs_payload, "kg")
```

### Task 4




##### Display average payload mass carried by booster version F9 v1.1


```python
# Filter rows where BoosterVersion is 'F9 v1.1' and calculate average PayloadMass
avg_payload_v1_1 = df[df['Booster_Version'] == 'F9 v1.1']['PAYLOAD_MASS__KG_'].mean()

print("Average payload mass for F9 v1.1:", avg_payload_v1_1, "kg")
```

### Task 5

##### List the date when the first succesful landing outcome in ground pad was acheived.


_Hint:Use min function_ 


```python
# Inspect unique Outcome values
print(df['Landing_Outcome'].unique())
```

```python
# Filter only successful ground pad landings
ground_success = df[df['Landing_Outcome'] == 'Success (ground pad)']

# Find the first successful landing date
first_success_date = ground_success['Date'].min()

print("First successful landing on ground pad:", first_success_date)
```

### Task 6

##### List the names of the boosters which have success in drone ship and have payload mass greater than 4000 but less than 6000


```python
# Filter for successful drone ship landings and payload mass between 4000 and 6000
boosters = df[
    (df['Landing_Outcome'] == 'Success (drone ship)') &
    (df['PAYLOAD_MASS__KG_'] > 4000) &
    (df['PAYLOAD_MASS__KG_'] < 6000)
]['Booster_Version'].unique()  # Get unique booster names

print("Boosters with successful drone ship landing and payload 4000-6000 kg:")
print(boosters)
```

### Task 7




##### List the total number of successful and failure mission outcomes


```python
# Define successful outcomes
success_outcomes = ['Success', 'Success (ground pad)', 'Success (drone ship)']

# Create a new column landing_class: 1 = success, 0 = failure
df['landing_class'] = df['Landing_Outcome'].apply(lambda x: 1 if x in success_outcomes else 0)

# Count total successes and failures
mission_counts = df['landing_class'].value_counts().rename({1: 'Success', 0: 'Failure'})
print(mission_counts)
```

### Task 8



##### List all the booster_versions that have carried the maximum payload mass, using a subquery with a suitable aggregate function.


```python
import pandasql as ps

query = """
SELECT "Booster_Version", "PAYLOAD_MASS__KG_"
FROM df
WHERE "PAYLOAD_MASS__KG_" = (
    SELECT MAX("PAYLOAD_MASS__KG_")
    FROM df
)
"""

result = ps.sqldf(query, locals())
print(result)
```

### Task 9


##### List the records which will display the month names, failure landing_outcomes in drone ship ,booster versions, launch_site for the months in year 2015.

**Note: SQLLite does not support monthnames. So you need to use  substr(Date, 6,2) as month to get the months and substr(Date,0,5)='2015' for year.**


```python
import pandasql as ps

query = """
SELECT 
    substr("Date", 6, 2) AS Month,
    "Landing_Outcome",
    "Booster_Version",
    "Launch_Site"
FROM df
WHERE substr("Date", 1, 4) = '2015'
  AND "Landing_Outcome" LIKE 'Failure (drone ship)%'
"""

result = ps.sqldf(query, locals())
print(result)
```

### Task 10




##### Rank the count of landing outcomes (such as Failure (drone ship) or Success (ground pad)) between the date 2010-06-04 and 2017-03-20, in descending order.


```python
import pandasql as ps

query = """
SELECT "Landing_Outcome", COUNT(*) AS outcome_count
FROM df
WHERE "Date" BETWEEN '2010-06-04' AND '2017-03-20'
GROUP BY "Landing_Outcome"
ORDER BY outcome_count DESC
"""

result = ps.sqldf(query, locals())
print(result)
```
